```python
# Importing NumPy
import numpy as np

# 1
arr_1d = np.array([1, 20, 19, 21, 4])
print("1D Array:\n", arr_1d)

arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("\n2D Array:\n", arr_2d)

# 2
print("\nArray Properties:")
print("Shape:", arr_2d.shape)
print("Size:", arr_2d.size)
print("Data Type:", arr_2d.dtype)

# 3
print("\nSlicing 1D Array:")
sliced_1d = arr_1d[1:4]
print(sliced_1d)

print("\nSlicing 2D Array:")
sliced_2d = arr_2d[:2, :2]
print(sliced_2d)

print("\nUsing Negative Indexing:")
neg_indexed = arr_1d[-3:]
print(neg_indexed)

# 4
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

print("\nArithmetic Operations:")
print("Addition:", arr1 + arr2)
print("Multiplication:", arr1 * 2)
print("Power:", arr1 ** 2)

# 5. 
mat1 = np.array([[1, 2], [3, 4]])
mat2 = np.array([[5, 6], [7, 8]])

print("\nMatrix Multiplication:")
print(np.dot(mat1, mat2))

# 6
zeros = np.zeros((2, 3))
ones = np.ones((3, 3))
identity = np.eye(3)
random_arr = np.random.rand(2, 3)

print("\nSpecial Arrays:")
print("Zeros:\n", zeros)
print("Ones:\n", ones)
print("Identity Matrix:\n", identity)
print("Random Array:\n", random_arr)

# 7
arr = np.array([1, 2, 3, 4, 5, 6])
reshaped = arr.reshape(2, 3)
flattened = reshaped.flatten()

print("\nReshaped Array:\n", reshaped)
print("Flattened Array:\n", flattened)

# 8
stats_arr = np.array([10, 20, 30, 40, 50])

print("\nStatistical Functions:")
print("Mean:", np.mean(stats_arr))
print("Median:", np.median(stats_arr))
print("Standard Deviation:", np.std(stats_arr))
print("Sum:", np.sum(stats_arr))
print("Min:", np.min(stats_arr))
print("Max:", np.max(stats_arr))
```

```python

import matplotlib.pyplot as plt
import numpy as np 


x = np.arange(1, 6)
y = x**2  

plt.figure(figsize=(6, 4))
plt.plot(x, y, marker='o', linestyle='-', color='b', label="y = x")


plt.title(' Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.legend()

# Show plot
plt.show()

plt.figure(figsize=(6, 4))
plt.plot(x, y, 'g--', marker='s', label="y = x")

# Adding titles, labels, and grid
plt.title(' Line Plot with Labels')
plt.xlabel('Input Values')
plt.ylabel('Squared Values')
plt.legend()
plt.grid(True)

plt.show()


categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

plt.figure(figsize=(6, 4))
plt.bar(categories, values, color=['red', 'blue', 'green', 'orange'])

# Customizing the bar chart
plt.title(' Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.show()


x_scatter = np.random.rand(20) * 10  # Random X values
y_scatter = np.random.rand(20) * 10  # Random Y values

plt.figure(figsize=(6, 4))
plt.scatter(x_scatter, y_scatter, color='purple', marker='x', s=100)


plt.title(' Scatter Plot Example')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)

plt.show()


data = np.random.randn(1000) 

plt.figure(figsize=(6, 4))
plt.hist(data, bins=20, color='cyan', edgecolor='black')


plt.title(' Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()


labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 30, 20, 10]
colors = ['gold', 'blue', 'green', 'red']

plt.figure(figsize=(6, 4))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)


plt.title(' Pie Chart Example')

plt.show()
```

```python
# Importing required libraries
import matplotlib.pyplot as plt

# Data for the two groups
group_A = [12, 15, 14, 13, 16, 18, 19, 15, 14, 20, 17, 14, 15, 40, 45, 50, 62]
group_B = [12, 17, 15, 13, 19, 20, 21, 18, 17, 16, 15, 14, 16, 15]

# Creating a figure with two subplots (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Box plot for Group A
axes[0].boxplot(group_A, patch_artist=True, boxprops=dict(facecolor='lightblue'))
axes[0].set_title(' Box Plot - Group A')
axes[0].set_ylabel('Measurement Values')

# Box plot for Group B
axes[1].boxplot(group_B, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
axes[1].set_title(' Box Plot - Group B')
axes[1].set_ylabel('Measurement Values')

# Adding an overall title
fig.suptitle('Comparison of Measurements: Group A vs Group B', fontsize=14)

# Display the plots
plt.show()
```
